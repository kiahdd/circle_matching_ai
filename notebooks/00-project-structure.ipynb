# CircleMatch Project Structure

This notebook describes the overall architecture, data model, and initial roadmap for the CircleMatch project.

It is intentionally code-free to keep the early design focused on concept, scope, and structure.

## Project overview

CircleMatch is an AI-powered community assistant that forms small groups of 4–6 people from free-text onboarding answers. The goal is to create emotionally coherent social circles that can actually meet, not just produce a list of matches.

## Key components

- User onboarding
- Embedding and trait representation
- Clustering and group formation
- Circle scoring and explanation
- Optional coordination layer for meetup suggestions

## Proposed folder structure

- `main.py`
- `utils/`
  - `__init__.py`
  - `data.py`
  - `matching.py`
  - `telebot.py`
- `notebooks/`
  - `00-project-structure.ipynb`

This structure supports a small prototype with clear separation between the orchestration layer, helper utilities, and research documentation.

## First-version goals

1. Collect free-text onboarding answers
2. Store user responses and derive embeddings
3. Form circles of 4–6 people
4. Generate a short explanation for each circle
5. Keep the prototype narrow and focused on matching logic

## Next steps

- choose an embedding strategy

## Current workflow split

The current implementation separates dataset creation from data analysis.

- Use `python main.py generate --n-batches 2 --n-users 25` to generate and save new datasets.
- Use `python main.py analyze --path datasets` to load existing saved datasets for analysis.

This split keeps dataset generation isolated from downstream embedding and clustering work.
